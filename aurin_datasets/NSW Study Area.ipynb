# NSW Study Area

This is an exercise to select our study area for the AusUrb-HI Heat Vulnerability Indicator, based on column and attribute names.

```python
%matplotlib inline
import geopandas as gpd
import pandas as pd
```

Import and inspect shapefiles

```python
sa1=gpd.read_file("Study Area/sa1_2016/SA1_2016_AUST.shp")
```

```python
sa2=gpd.read_file("Study Area/sa2_2016/SA2_2016_AUST.shp")
```

```python
sa2
```

```python
sa2.plot(figsize=(10,10))
```

## SA2 Areas

We are looking for SA2 areas that are within Greater Sydney, Tweed Valley, Newcastle-Maitland, Wollongong and Albury areas. We will separate regional and capital areas, then concatenate them in the end. 

```python
sa2_regional = sa2.loc[sa2['SA3_NAME16'].isin(['Tweed Valley', 'Newcastle', 'Maitland', 'Wollongong', 'Albury']), :]
```

```python
sa2_regional
```

```python
sa2_capital = sa2.loc[sa2['GCC_NAME16']=='Greater Sydney']
```

```python
sa2_capital
```

```python
sa2_capital.plot()
```

```python
sa2_regional.count()
```

```python
sa2_capital.count()
```

```python
sa2_all = pd.concat([sa2_regional, sa2_capital])
sa2_all.count()
```

```python
sa2_all.tail(20)
```

```python
sa2_all.plot(figsize=(20,20))
```

## Export shapefiles

```python
sa2_all.to_file("Study Area/sa2_nsw.shp")
```

## Boundaries

Dissolving geometries to obtain boundaries of the study areas. These will be helpful when extracting raster data and pointclouds.

```python
sa2_diss = sa2_all.dissolve()
```

```python
sa2_diss
```

```python
sa2_diss.plot(figsize=(20,20))
```

```python
sa2_diss.to_file("Study Area/boundaries.shp")
```
